**K-Means Clustering**

The K-Means algorithm is one of the fastest and simplest clustering algorithms. It starts by initializing centroids randomly, typically by selecting distinct instances from the dataset and placing the centroids at their locations. The algorithm then repeats the following steps until convergence (when the centroids stop moving): each instance in the dataset is assigned to the closest centroid, and the centroids are updated to be the mean of the instances assigned to them

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Define the new centers of the blobs
new_blob_centers = np.array([
    [1.0, 3.0],   # Center of the first blob
    [-2.0, 2.0],  # Center of the second blob
    [-3.0, 4.0],  # Center of the third blob
    [-1.0, 5.0],  # Center of the fourth blob
    [2.0, 1.0]    # Center of the fifth blob
])

# Define the new standard deviation of the blobs
new_blob_std = np.array([0.5, 0.6, 0.2, 0.3, 0.4])

# Generate synthetic data with the new blob centers and standard deviations
X, y = make_blobs(n_samples=2000, centers=new_blob_centers, cluster_std=new_blob_std, random_state=42)

# Define the number of clusters for K-means
k = 5

# Create a KMeans instance with k clusters
kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)

# Fit the KMeans model and predict the cluster for each data point
y_pred = kmeans.fit_predict(X)

# Function to plot clusters
def plot_clusters(X, y=None):
    plt.scatter(X[:, 0], X[:, 1], c=y, s=1)
    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$", rotation=0)


# Plotting the clusters
plt.figure(figsize=(8, 4))
plot_clusters(X)
plt.gca().set_axisbelow(True)
plt.grid()
plt.show()
```
